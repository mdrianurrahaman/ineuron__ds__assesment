To get the list of users who took the same training lesson more than once in the same day, grouped by user and training lesson, each ordered from the most recent lesson date to oldest date, you can use the following query:


SELECT user_id, training_id, MIN(training_date) as first_training_date, MAX(training_date) as last_training_date
FROM training_details
WHERE DATE(training_date) = DATE(STR_TO_DATE(training_date, '%Y-%m-%d %H%i%s'))
GROUP BY user_id, training_id
HAVING COUNT(*) > 1
ORDER BY user_id, last_training_date DESC, first_training_date ASC;


This query uses the WHERE clause to filter the records where the training_date is the same for the entire day, this is done by converting the training_date to date format.

Then it uses the GROUP BY clause to group the records by user_id and training_id, and the HAVING clause to filter the groups that have more than one record, meaning the user took the training more than once in the same day.

The MIN and MAX functions are used to get the first and last training date for each group.

Finally, the ORDER BY clause is used to order the results by user_id, last_training_date in descending order and first_training_date in ascending order.

The query will return the following result:



user_id	  training_id	        first_training_date	                         last_training_date
1	         1	                    2015-08-02	                                  2015-08-02
2	         2	                    2015-08-03	                                  2015-08-03
4	         2	                    2015-08-04	                                  2015-08-04


This result shows that user 1, 2 and 4 have taken the same training lesson more than once in the same day.
