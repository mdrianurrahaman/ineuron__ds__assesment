# Answer of question no 1 :

The query SELECT * FROM runners WHERE id NOT IN (SELECT winner_id FROM races) will return the rows from the runners table where the id is not present in the winner_id column of the races table.

In this case, the result of the query will be:

id	name
1	John Doe
3	Alice Jones
5	Lisa Romero


The id of John Doe (1) and Alice Jones (3) are not present in the winner_id column of the races table, and Lisa Romero's id is not present in the winner_id column of the races table. Therefore, their rows are returned by the query.

The issue with this query is that it will not return any rows where the id in the runners table is NULL, even if there is no corresponding winner_id in the races table.

An alternative version of this query that avoids this issue is:

SELECT * FROM runners WHERE id IS NOT NULL AND id NOT IN (SELECT winner_id FROM races WHERE winner_id IS NOT NULL)

This query first filters out any rows in the runners table where the id is NULL, and then checks for rows where the id is not present in the winner_id column of the races table, only when the winner_id is not null. This way, it will return all the rows that are not present in the winner_id column of the races table, even when the id in runners table is null.
