To fetch the values in test_a that are present in test_b, you can use the INTERSECT keyword, but as you've specified not to use the NOT keyword, we can use LEFT JOIN to find the values that are present in both tables.


Here is the query:

SELECT test_a.id
FROMa
JOIN test_b
ON test_a.id = test_b.id;


This query will return the values 10 and 50, which are present in both test_a and test_b.

To fetch the values in test_a that are not present in test_b, you can use the LEFT JOIN again and check for the NULL values in the test_b.id column.

Here is the query:

SELECT test_a.id
FROM test_a
LEFT JOIN test_b
ON test_a.id = test_b.id
WHERE test_b.id IS NULL;


This query will return the values 20 and 40, which are present in test_a but not in test_b.

By using LEFT JOIN we can find the matching values as well as the non-matching values between two tables.

Note: In case you want to fetch the non-matching values only, you can use MINUS keyword instead of LEFT JOIN
