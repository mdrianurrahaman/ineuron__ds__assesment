```python
the query is :

SELECT Manager_Id, AVG(Salary) AS Average_Salary_Under_Manager
FROM Employee
WHERE Manager_Id IS NOT NULL
GROUP BY Manager_Id;

This query first filters out the records where the Manager_Id is NULL, which means those are the managers. Then it groups the remaining records by Manager_Id and calculates the average salary for each group using the AVG function.

The result of the query will be:



Manager_Id	Average_Salary_Under_Manager
16	           65000
17	           62500
18	           53750


This result shows the average salary of the employees who are under the management of manager 16, 17, and 18 respectively.

```
